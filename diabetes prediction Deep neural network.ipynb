```python
train_ds=TensorDataset(inputs,targets)
```

```python
train_ds,test_ds=random_split(train_ds,[600,168])
```

```python
batch_size=20
train_dl=DataLoader(train_ds,batch_size)
```

```python
class Dnn(nn.Module):
    def __init__(self):
        super().__init__()
        self.linear1=nn.Linear(8,16)
        self.linear2=nn.Linear(16,16)
        self.linear3=nn.Linear(16,32)
        self.linear4=nn.Linear(32,32)
        self.linear5=nn.Linear(32,8)
        self.linear6=nn.Linear(8,8)
        self.linear7=nn.Linear(8,2)
    def forward(self,xb):
        out=self.linear1(xb)
        out=F.relu(out)
        out=self.linear2(out)
        out=F.relu(out)
        out=self.linear3(out)
        out=F.relu(out)
        out=self.linear4(out)
        out=F.relu(out)
        out=self.linear5(out)
        out=F.relu(out)
        out=self.linear6(out)
        out=F.relu(out)
        pred=self.linear7(out)
        return pred
    def training_step(self,batch):
        inpt,targ=batch
        pred=self(inpt)
        loss=F.cross_entropy(pred,targ)
        acc_val = (torch.argmax(pred, dim=1) == targ).float().mean().item()
        print(f'loss-->{loss}||accuracy-->{acc_val}')
        return loss
```

```python
def fit(epochs,lr,model,train_dl):
    optimizer=torch.optim.Adam(model.parameters(),lr)
    for epoch in range(epochs):
        print(epoch)
        for batch in train_dl:
            loss=model.training_step(batch)
            optimizer.zero_grad()
            loss.backward()
            optimizer.step()
```

```python
model=Dnn()
```

```python
fit(100,0.01,model,train_dl)
```
